## 1. Análisis de Desempeño Logístico

**Situación:** Evaluación de envíos según tipo de transporte (Terrestre, Aéreo, Marítimo).

### Probabilidades Básicas:
- **Éxito Global ($P(S)$):** De un total de $n=350$ pedidos, $250$ fueron exitosos.
  $$P(S) = \frac{250}{350} = \frac{5}{7} \approx 0.7143$$

- **Aéreo dado Éxito ($P(A|S)$):** Conocido un envío exitoso, ¿cuál es su origen aéreo?
  $$P(A|S) = \frac{P(S \cap A)}{P(S)} = \frac{80}{250} = 0.32$$

- **Fallo en Marítimo ($P(F|M)$):** Riesgo específico del transporte por mar.
  $$P(F|M) = \frac{50}{100} = 0.5$$

### Ley de Probabilidad Total:
Verificamos que la probabilidad de éxito dependa de la suma ponderada de los canales:
$$P(S) = (120/150) \cdot (150/350) + (80/100) \cdot (100/350) + (50/100) \cdot (100/350) = 250/350 = 0.7143$$


## 2. Monitoreo Poisson y Análisis Bayesiano

**Datos (15 hrs):** `0, 1, 2, 3, 2, 1, 0, 2, 3, 4, 2, 1, 2, 3, 2`.

### Estadísticas Muestrales:
- **Media ($\bar{x}$):** $1.8667$
- **Varianza ($s^2$):** $1.2667$
(La cercanía relativa entre media y varianza sugiere que el modelo de Poisson es adecuado para bajas tasas de eventos).

### Inferencia de Causa (Bayes):
Existen dos servidores (A: $\lambda=2$, p=65%; B: $\lambda=1$, p=35%). Si ocurre $k=2$ fallos en una hora:
- $P(2|A) = \frac{e^{-2}2^2}{2!} \approx 0.2707$
- $P(2|B) = \frac{e^{-1}1^2}{2!} \approx 0.1839$
**Probabilidad de que sea el Servidor A:**
$$P(A|2) = \frac{P(2|A)P(A)}{P(2|A)P(A) + P(2|B)P(B)} \approx \frac{0.1759}{0.1759 + 0.0644} = 0.7321$$


## 3. Variable Continua y Funciones Generadoras

Dada $f(x) = xe^{-x}$ para $x > 0$:

### Función Generadora de Momentos (MGF)
$$M_X(t) = \int_0^\infty e^{tx} x e^{-x} dx = \int_0^\infty x e^{-(1-t)x} dx = \frac{1!}{(1-t)^2} = (1-t)^{-2}, \quad t < 1$$

### Media Teórica
$$\mathbb{E}[X] = M_X'(0) = [-2(1-t)^{-3} \cdot (-1)]_{t=0} = 2$$


## 4. Análisis de Tiempos Académicos (Normal)

Basado en 20 observaciones:
- Media $\mu = 17.26$ min.
- Desviación $\sigma = 2.60$ min.

### Probabilidades e Indicadores:
- **Rango Intermedio ($16 < X < 19$):**
  $$P(X < 19) - P(X < 16) \approx 0.748 - 0.314 = 0.434$$

- **Tiempo de Referencia (Percentil 95):**
  Para asegurar que el 95% de los estudiantes terminen, se debe establecer un tiempo de:
  $$x = \mu + z_{0.95}\sigma \approx 17.26 + 1.645(2.60) = 21.54 \text{ minutos.}$$
