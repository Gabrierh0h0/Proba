## Ejercicio 1: Probabilidad Condicional y Teorema de Bayes

**Enunciado:**
En una caja hay 5 bolas negras (N) y 4 blancas (B). En una segunda caja hay 3 negras y 2 blancas. Se saca una bola de la primera caja y se deposita en la segunda. Luego se saca una bola de la segunda caja.

### Datos:
- Caja 1: $5N, 4B$. Total = 9.
- Caja 2: $3N, 2B$. Total = 5.

Sean:
- $N_1$: Transferir bola negra.
- $B_1$: Transferir bola blanca.
- $B_2$: Sacar bola blanca de la caja 2.

Probabilidades iniciales:
- $P(N_1) = 5/9$
- $P(B_1) = 4/9$

Si transferimos una negra ($N_1$), la caja 2 queda con $4N, 2B$ (Total 6).
- $P(B_2 | N_1) = 2/6 = 1/3$

Si transferimos una blanca ($B_1$), la caja 2 queda con $3N, 3B$ (Total 6).
- $P(B_2 | B_1) = 3/6 = 1/2$

### a) Probabilidad de que la segunda bola sea blanca ($P(B_2)$):
Usando la Ley de Probabilidad Total:
$$
P(B_2) = P(B_2 | N_1)P(N_1) + P(B_2 | B_1)P(B_1)
$$
$$
P(B_2) = \left(\frac{1}{3}\right)\left(\frac{5}{9}\right) + \left(\frac{1}{2}\right)\left(\frac{4}{9}\right)
$$
$$
P(B_2) = \frac{5}{27} + \frac{4}{18} = \frac{5}{27} + \frac{2}{9} = \frac{11}{27} \approx 0.4074
$$

### b) Si la segunda fue blanca, ¿probabilidad de que se transfirió una negra? ($P(N_1 | B_2)$):
Usando el Teorema de Bayes:
$$
P(N_1 | B_2) = \frac{P(B_2 | N_1)P(N_1)}{P(B_2)} = \frac{5/27}{11/27} = \frac{5}{11} \approx 0.4545
$$


## Ejercicio 2: Variable Aleatoria Continua y MGF

**Enunciado:**
Función de densidad: $f(x) = 4xe^{-2x}$ para $x > 0$.

### Función Generatriz de Momentos $M_X(t)$:
$$
M_X(t) = E[e^{tX}] = \int_0^\infty e^{tx} (4xe^{-2x}) dx = 4 \int_0^\infty x e^{-(2-t)x} dx
$$
Usando la fórmula $\int_0^\infty x^n e^{-ax} dx = \frac{n!}{a^{n+1}}$ con $n=1$ y $a=2-t$:
$$
M_X(t) = 4 \cdot \frac{1!}{(2-t)^2} = \frac{4}{(2-t)^2} = (1 - t/2)^{-2}, \quad t < 2
$$

### Cálculo de Media y Varianza:
1. Derivada de $M_X(t)$:
$$
M_X'(t) = -2(1 - t/2)^{-3} \cdot (-1/2) = (1 - t/2)^{-3}
$$
$$E[X] = M_X'(0) = (1 - 0)^{-3} = 1$$

2. Segunda derivada:
$$
M_X''(t) = -3(1 - t/2)^{-4} \cdot (-1/2) = \frac{3}{2}(1 - t/2)^{-4}
$$
$$E[X^2] = M_X''(0) = 1.5$$

**Varianza:**
$$
\sigma^2 = E[X^2] - (E[X])^2 = 1.5 - 1^2 = 0.5
$$


## Ejercicio 3: Distribuciones Discretas

**Datos:** $p = 0.35$.

### a) $n=15$, al menos 7 éxitos:
$$
P(X \ge 7) = 1 - P(X \le 6) = 1 - \sum_{k=0}^{6} \binom{15}{k} (0.35)^k (0.65)^{15-k} \approx 0.2452
$$

### b) Quiebra si el primer éxito no aparece antes de la quinta instalación:
Esto ocurre si hay 4 fracasos seguidos:
$$
P(\text{Quiebra}) = (0.65)^4 \approx 0.1785
$$

### c) Éxito absoluto si el tercer éxito aparece antes de la novena instalación:
Es equivalente a tener al menos 3 éxitos en 8 ensayos:
$$
P(Y_3 \le 8) = P(X_8 \ge 3) = 1 - P(X_8 \le 2) \approx 0.5722
$$
