```python
from scipy.stats import expon as exp 
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = exp.cdf(2, scale=1/0.5)
b = exp.cdf(6, scale=1/0.5)

b-a
```

```python
# Parámetros (lambda = 0.5, scale = 2)
lambda_val = 0.5
scale = 1 / lambda_val

# Datos para la curva
x = np.linspace(0, 15, 1000)
y = exp.pdf(x, scale=scale)

plt.figure(figsize=(10, 5))
plt.plot(x, y, color='#2c3e50', lw=2.5, label=f'PDF Exponencial (λ={lambda_val})')

# Sombreado del área calculada P(2 < X < 6)
x_fill = np.linspace(2, 6, 200)
y_fill = exp.pdf(x_fill, scale=scale)
plt.fill_between(x_fill, y_fill, color='#3498db', alpha=0.4, label='P(2 < X < 6)')

# Estética
plt.title('Distribución Exponencial: Visualización de Probabilidad', fontsize=14, pad=15)
plt.xlabel('X', fontsize=12)
plt.ylabel('Densidad de Probabilidad', fontsize=12)
plt.axvline(2, color='red', linestyle='--', alpha=0.5)
plt.axvline(6, color='red', linestyle='--', alpha=0.5)
plt.legend(frameon=True, facecolor='white')
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()
```
