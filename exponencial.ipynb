# Distribuciones Continuas y Momentos (Exponencial/Gamma)

Este cuaderno muestra cómo trabajar con variables aleatorias continuas, el cálculo de sus Funciones Generadoras de Momentos (MGF) y la obtención de estadísticos básicos como la media y varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
from scipy.integrate import quad
```

## 1. Análisis de una Función de Densidad $X$

Dada la función de densidad:
$$f(x) = xe^{-x}, \quad x > 0$$

### a) Derivación de la MGF
La MGF se define como $M_X(t) = E[e^{tX}]$. Aplicando la integral:
$$M_X(t) = \int_0^\infty e^{tx} x e^{-x} dx = \int_0^\infty x e^{-(1-t)x} dx = (1-t)^{-2}, \quad t < 1$$

### b) Cálculo de la Media a través de Momentos
La media $\mathbb{E}[X]$ es la primera derivada de la MGF evaluada en $t=0$.
$$M_X'(t) = 2(1-t)^{-3} \implies \mathbb{E}[X] = M_X'(0) = 2$$

```python
# Definimos la función f(x) = x * exp(-x)
def pdf(x):
    return x * np.exp(-x)

# Verificamos numéricamente la media E[X] = integral de x*f(x)
mean_val, _ = quad(lambda x: x * pdf(x), 0, np.inf)
print(f"Media calculada (numérica): {mean_val:.4f}")
```

## 2. Visualización de la Densidad
Podemos observar el comportamiento de la variable $X$ en el tiempo o espacio.

```python
x_vals = np.linspace(0, 10, 1000)
y_vals = pdf(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, color='#45B39D', lw=2.5)
plt.fill_between(x_vals, y_vals, alpha=0.3, color='#45B39D')
plt.axvline(mean_val, color='orange', linestyle='--', label=f'Media ({mean_val:.1f})')
plt.title('Densidad de Probabilidad: $f(x) = xe^{-x}$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
