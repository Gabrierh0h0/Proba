# Distribución Normal en Python

Este cuaderno para **Google Colab** muestra cómo trabajar la **distribución normal**: probabilidades, percentiles, media, varianza, simulación, histograma y explicaciones de lo que hace el código.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## 1. Idea básica

La distribución normal es continua y se usa para modelar variables que tienden a agruparse alrededor de un promedio.

Si $X\sim N(\mu,\sigma)$, entonces $\mu$ es la media y $\sigma$ la desviación estándar.


```python
mu = 50
sigma = 8
print('Media =', mu)
print('Desviación estándar =', sigma)
```

## 2. Probabilidades

Como la normal es continua, las probabilidades se calculan sobre intervalos. En una variable continua, $P(X=a)=0$ para un punto exacto.


```python
a = 45
b = 60
print(f'P(X <= {a}) = {norm.cdf(a, mu, sigma):.6f}')
print(f'P(X >= {a}) = {norm.sf(a, mu, sigma):.6f}')
print(f'P({a} <= X <= {b}) = {norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma):.6f}')
```

## 3. Percentiles

La función `ppf` devuelve cuantiles o percentiles.


```python
percentil_90 = norm.ppf(0.90, mu, sigma)
print(f'Percentil 90 = {percentil_90:.4f}')
```

## 4. Media y varianza

Para la normal:

$$\mathbb{E}[X]=\mu,\qquad \mathrm{Var}(X)=\sigma^2.$$

```python
media = norm.mean(mu, sigma)
varianza = norm.var(mu, sigma)
print(f'Media teórica    = {media:.4f}')
print(f'Varianza teórica = {varianza:.4f}')
```

## 5. Simulación


```python
np.random.seed(42)
muestra = np.random.normal(loc=mu, scale=sigma, size=1000)
print('Primeros 10 datos simulados:')
print(muestra[:10])
```

```python
print(f'Media muestral    = {np.mean(muestra):.4f}')
print(f'Varianza muestral = {np.var(muestra):.4f}')
```

## 6. Histograma con densidad teórica


```python
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 400)
densidad = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10,6))
plt.hist(muestra, bins=30, density=True, alpha=0.7, edgecolor='black')
plt.plot(x, densidad, linewidth=2)
plt.axvline(mu, linestyle='--', linewidth=2)
plt.xlabel('x')
plt.ylabel('Densidad')
plt.title(f'Histograma y densidad normal teórica N({mu}, {sigma})')
plt.grid(alpha=0.3)
plt.show()
```

La línea punteada marca la media. La forma simétrica alrededor de la media es una característica esencial de la distribución normal.


## 7. Curva de densidad


```python
plt.figure(figsize=(10,6))
plt.plot(x, densidad, linewidth=2)
plt.axvline(mu, linestyle='--', linewidth=2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Densidad de una Normal({mu}, {sigma})')
plt.grid(alpha=0.3)
plt.show()
```

## 8. Función reutilizable


```python
def resumen_normal(mu, sigma, a, b):
    print(f'Distribución: Normal(mu={mu}, sigma={sigma})')
    print(f'P(X <= {a}) = {norm.cdf(a, mu, sigma):.6f}')
    print(f'P(X >= {a}) = {norm.sf(a, mu, sigma):.6f}')
    print(f'P({a} <= X <= {b}) = {norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma):.6f}')
    print(f'Media       = {norm.mean(mu, sigma):.4f}')
    print(f'Varianza    = {norm.var(mu, sigma):.4f}')

resumen_normal(70, 12, 65, 80)
```
