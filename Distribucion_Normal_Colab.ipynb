# Distribución Normal en Python

Este cuaderno para **Google Colab** muestra cómo trabajar la **distribución normal**: probabilidades, percentiles, media, varianza, simulación, histograma y explicaciones de lo que hace el código.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## 1. Idea básica

La distribución normal es continua y se usa para modelar variables que tienden a agruparse alrededor de un promedio.

Si $X\sim N(\mu,\sigma)$, entonces $\mu$ es la media y $\sigma$ la desviación estándar.


```python
mu = 50
sigma = 8
print('Media =', mu)
print('Desviación estándar =', sigma)
```

## 2. Probabilidades

Como la normal es continua, las probabilidades se calculan sobre intervalos. En una variable continua, $P(X=a)=0$ para un punto exacto.


```python
a = 45
b = 60
print(f'P(X <= {a}) = {norm.cdf(a, mu, sigma):.6f}')
print(f'P(X >= {a}) = {norm.sf(a, mu, sigma):.6f}')
print(f'P({a} <= X <= {b}) = {norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma):.6f}')
```

## 3. Percentiles

La función `ppf` devuelve cuantiles o percentiles.


```python
percentil_90 = norm.ppf(0.90, mu, sigma)
print(f'Percentil 90 = {percentil_90:.4f}')
```

## 4. Media y varianza

Para la normal:

$$\mathbb{E}[X]=\mu,\qquad \mathrm{Var}(X)=\sigma^2.$$

```python
media = norm.mean(mu, sigma)
varianza = norm.var(mu, sigma)
print(f'Media teórica    = {media:.4f}')
print(f'Varianza teórica = {varianza:.4f}')
```

## 5. Simulación


```python
np.random.seed(42)
muestra = np.random.normal(loc=mu, scale=sigma, size=1000)
print('Primeros 10 datos simulados:')
print(muestra[:10])
```

```python
print(f'Media muestral    = {np.mean(muestra):.4f}')
print(f'Varianza muestral = {np.var(muestra):.4f}')
```

## 6. Histograma con densidad teórica


```python
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 400)
densidad = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10,6))
plt.hist(muestra, bins=30, density=True, alpha=0.7, edgecolor='black')
plt.plot(x, densidad, linewidth=2)
plt.axvline(mu, linestyle='--', linewidth=2)
plt.xlabel('x')
plt.ylabel('Densidad')
plt.title(f'Histograma y densidad normal teórica N({mu}, {sigma})')
plt.grid(alpha=0.3)
plt.show()
```

La línea punteada marca la media. La forma simétrica alrededor de la media es una característica esencial de la distribución normal.


## 7. Caso de Estudio: Tiempos en Tareas Académicas

**Datos Observados:** Se registran los tiempos (en minutos) que tardan 20 estudiantes en resolver un ejercicio de cálculo.

### a) Análisis Descriptivo y Normatividad

```python
tiempos = np.array([12.1, 13.5, 14.2, 15.0, 15.3, 15.8, 16.1, 16.4, 16.8, 17.0, 17.3, 17.6, 18.0, 18.4, 18.9, 19.2, 19.8, 20.5, 21.0, 22.3])
media_t = np.mean(tiempos)
desviacion_t = np.std(tiempos, ddof=1)
print(f"Media: {media_t:.2f} min")
print(f"Desviación estándar: {desviacion_t:.2f} min")
```

**Justificación:** Los datos presentan una distribución equilibrada alrededor de la media y una dispersión moderada, lo cual hace razonable el uso de un modelo normal para representar el comportamiento del grupo.

### b) Probabilidades en el Rango [16, 19]

```python
prob_intervalo = norm.cdf(19, media_t, desviacion_t) - norm.cdf(16, media_t, desviacion_t)
print(f"P(16 < X < 19): {prob_intervalo:.4f}")
```

### c) Tiempo de Referencia para la Mayoría
Determinar un tiempo adecuado tal que la gran mayoría (ej. el 95%) logre terminar.

```python
tiempo_referencia = norm.ppf(0.95, media_t, desviacion_t)
print(f"Tiempo sugerido (P95): {tiempo_referencia:.2f} minutos")
```

**Interpretación:** Si establecemos un tiempo de aproximadamente **21.5 minutos**, podemos garantizar que la gran mayoría de los estudiantes (95%) tendrá tiempo suficiente para completar el ejercicio.
