```python
a= wb.cdf(4, c=2, scale=5)
b= 1 - wb.cdf(6, c=2, scale=5)
c= wb.ppf(0.9, c=2, scale=5)
e= wb.mean(c=2, scale=5)
f= wb.var(c=2, scale=5)
falla = wb.pdf(4, c=1/2, scale=1/4) / wb.sf(4, c=1/2, scale=1/4)

print(a)
print(b)
print(c)
print(e)
print(f)
print(falla)
```
